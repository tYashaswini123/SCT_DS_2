**Performing data cleaning and exploratory data analysis (EDA) on Titanic Dataset**

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: File Upload in Google Colab
from google.colab import files

# Prompt user to upload files
print("Please upload your train.csv file:")
uploaded_train = files.upload()
print("Please upload your test.csv file:")
uploaded_test = files.upload()
print("Please upload your gender_submission.csv file:")
uploaded_gender_submission = files.upload()

# Step 2: Load the uploaded files into pandas DataFrames
train = pd.read_csv(next(iter(uploaded_train)))
test = pd.read_csv(next(iter(uploaded_test)))
gender_submission = pd.read_csv(next(iter(uploaded_gender_submission)))

# Step 3: Display basic information about the datasets
print("Train dataset overview:")
print(train.info())
print("\nTest dataset overview:")
print(test.info())
print("\nGender submission dataset overview:")
print(gender_submission.info())

# Step 4: Data Cleaning for Train Dataset
# Handle missing values
train['Age'].fillna(train['Age'].median(), inplace=True)
train['Embarked'].fillna(train['Embarked'].mode()[0], inplace=True)
train.drop(columns=['Cabin'], inplace=True)

# Check for remaining missing values
print("\nMissing values in train dataset after cleaning:")
print(train.isnull().sum())

# Step 5: Exploratory Data Analysis (EDA)
# Survival count plot
plt.figure(figsize=(8, 4))
sns.countplot(data=train, x='Survived')
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()

# Survival by gender
plt.figure(figsize=(8, 4))
sns.countplot(data=train, x='Survived', hue='Sex')
plt.title('Survival by Gender')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()

# Survival by passenger class
plt.figure(figsize=(8, 4))
sns.countplot(data=train, x='Survived', hue='Pclass')
plt.title('Survival by Passenger Class')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.legend(title='Pclass')
plt.show()

# Age distribution with survival
plt.figure(figsize=(8, 4))
sns.histplot(data=train, x='Age', kde=True, bins=30, hue='Survived')
plt.title('Age Distribution with Survival')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Display a summary of data cleaning and analysis
print("\nData Cleaning and EDA Complete.")
```
